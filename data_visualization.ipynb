```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from plotly.offline import init_notebook_mode
```

```python
data = pd.read_csv('Data/data_train.csv')
```

```python
data.describe()
```

```python
# Jupyter-friendly Plotly init (works well in classic Jupyter / many notebook setups)
init_notebook_mode(connected=True)

# If you have non-numeric cols (strings, categories), this prevents Plotly errors
num = data.select_dtypes(include="number")

for col in num.columns:
    fig = go.Figure(go.Scatter(x=num.index, y=num[col], mode="lines", name=str(col)))
    fig.update_layout(template="plotly_white", title=str(col), xaxis_title="index", yaxis_title=col)

    fig.update_layout(
        template="plotly_white",
        title="All numeric columns",
        xaxis_title="index",
        yaxis_title="value",
        legend_title="column",
        height=500,
    )

    fig.show()
```
